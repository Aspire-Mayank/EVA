* **Method which return image in matrix format from URL which helps easy calculation with filters.**

```
#Import necessary pacakges
import numpy as np
import cv2
from matplotlib import pyplot as plt
from urllib.request import urlopen 

#Method #1 
def url_to_image(url):
  resp = urlopen(url)
  image= np.asarray(bytearray(resp.read()), dtype='uint8')
  image =cv2.imdecode(image,cv2.IMREAD_COLOR)
  
  #return the image
  return image
  
```

```
#Calling url to image Method which convert image to array/matrix.
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

Original Input Image from URL without applying any mask/filter.
this input image has three channels RGB(colored)

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

**Applied Canny EDGE Detection Algorithm :** To get edges of Input image from OpenCV package.

```
edges = cv2.Canny(image,100,200)
cv2_imshow(edges)
```

* **Vertical Edge Detection :** This is Sobel Operator for y-axis
 1. Filter/Mask 3*3 Applied to 2D Input image. 

```
kernal = np.float32([[-1,0,1],
                    [-1,0,1],
                    [-1,0,1]])
dst = cv2.filter2D(image,-1,kernal)
cv2_imshow(dst)
```

* **Horizontal Edge Detection :** This is Sobel Operator for x-axis
 1. Filter/Mask 3*3 Applied to 2D Input image. 

```
#kernal = np.ones((3,3),np.float32)/25
kernal = np.float32([[-1,-1,-1],
                     [0,0,0],
                     [1,1,1]])
dst = cv2.filter2D(image,-1,kernal)
cv2_imshow(dst)
```

* **Blur Image filter :**  Its gussian filter/mask which blur the input content in image.
 1. Filter/Mask 3*3 Applied to Input image. 

```
kernal = np.float32([[.1,.2,.1],
                    [.2,.4,.2],
                    [.1,.2,.1]])
dst = cv2.filter2D(image, -1,kernal)
cv2_imshow(dst)
```

* **45* Angle Edge Detection :**  It detects diagonal from x & y axis as both content includes in result post masking.
 1. Filter/Mask 3*3 Applied to Input image. 

```
kernal = np.float32([[-1,-.5,0],
                    [-.5,0,.5],
                    [0,.5,1]])
dst = cv2.filter2D(image, -1,kernal)
cv2_imshow(dst)
```

* **Sharpen Image :**  It enhance the image content as result post masking.
 1. Filter/Mask 3*3 Applied to Input image. 

```
kernal = np.float32([[0,-.2,0],
                    [-.2,1,-.2],
                    [0,-.2,0]])
dst = cv2.filter2D(image,-1,kernal)
cv2_imshow(dst)
```

* **Identity Mask/Function:**  It gives same results once applied to input content/image.
 1. Filter/Mask 3*3 Applied to 2D Input image

```
kernal = np.float32([[0,0,0],
                    [0,1,0],
                    [0,0,0]])
dst = cv2.filter2D(image,-1,kernal)
cv2_imshow(dst)
```
